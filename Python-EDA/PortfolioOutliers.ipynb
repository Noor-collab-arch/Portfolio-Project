# **APPlying Z-Score method on features Usage and Fitness as there std is almost 1**

 Z score = (x -mean) / std. deviation

 **Findings:**

*    No outliers found in Fitness feature
*   outlier in usage feature "dataset is [7, 7]" which are contextual Outliers



```python
z = (data_df['Fitness']-data_df['Fitness'].mean()/data_df['Fitness'].std())
#print(z)
#print(z.max())
#print(z.min())
```

```python
threshold = 3
outlier = []
for i in data_df['Fitness']:
    z = (i-data_df['Fitness'].mean())/data_df['Fitness'].std()
    if z > threshold or z < -threshold:
        outlier.append(i)

print('outlier in dataset is', outlier)
```

# **IQR based Filtration on Feature Miles**



*  75% data is around 114.75 and max value is 360 which mean we have to use iqr method to detect outlier

**Findings**


*   It has been observed in below code that in outliers maximum value of miles is 360 miles per week and minimum miles per week is 188 miles which is impossible to cover in a week by any customer so i deleted these outliers





```python
# calculating IQR
quartile_25 = data_df['Miles'].quantile(.25)
quartile_75 = data_df['Miles'].quantile(.75)
iqr = quartile_75 - quartile_25
print(iqr)

# Finding upper and lower limit
#lower limit = Q1-1.5*IQR ,   Upper limit = Q3+1.5*IQR
lower_limit= quartile_25 - 1.5 * iqr
Upper_limit = quartile_75 + 1.5 * iqr

print(lower_limit)
print(Upper_limit)

```

```python
#outliers
outlier_df = data_df[(data_df['Miles'] > Upper_limit) | (data_df['Miles']< lower_limit)]
outlier_df
```

**Max and Min value of Miles in Outliers**

```python
#outlier_df.shape
#outlier_df.describe()

# max value of outlier
outlier_df['Miles'].max()
```

```python
#min value of outlier
outlier_df['Miles'].min()
```

**Without Outliers**

```python
without_outlier_df= data_df[(data_df['Miles'] >= lower_limit) & (data_df['Miles'] <= Upper_limit)]
without_outlier_df
```

**Comparison Before Outliers and After Outliers**

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.boxplot(data_df['Miles'],  color= 'grey')
plt.title('Before Outliers')

plt.subplot(2,2,2)
sns.boxplot(without_outlier_df['Miles'], color= 'grey')
plt.title('After Outliers')

plt.show()
```
