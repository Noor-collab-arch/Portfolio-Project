**Feature Engineering:**
* Created another coloumn named as age_group to divide our customers into groups

```python
#function for age group
def age_group(x):
  if x < 18:
    return "Below 18"
  elif x >=18 and x<=30:
    return "18--30"
  elif x >30 and x<=40:
    return "30--40"
  elif x >40 and x<=50:
    return "40--50"
  elif x >50 and x<=60:
    return "50--60"
  else:
    return "above 60"
```

```python
# new column
data_df['age_group']=data_df['Age'].apply(age_group)
data_df.head(10)
```

```python
sns.countplot(data=data_df, x="Product", hue="age_group" ,  palette="viridis")
plt.title("Prodcut use By Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(title="Use by Age")
plt.show()
```

**Feature Engineering**: Created new column of income group

```python
# dividing into low,medium and high catagories by salary threshold
def income_group(x):
  if x>=29562 and x<=45000:
    return 'Low Income'
  elif x>45000 and x<=70000:
    return 'Medium Income'
  else:
    return 'High Income'
```

```python
# making new column for income group by applying function
data_df['Income group']=data_df['Income'].apply(income_group)
data_df.tail(10)
```

```python
#frequency table Product vs Income group
product_income=pd.crosstab(index=data_df['Product'], columns=data_df['Income group'],margins=True)
print(product_income)
```
